```python
from SBM_attributed import *
import cvxpy as cp
import networkx as nx
from scipy.optimize import minimize,LinearConstraint
from sklearn.metrics import rand_score, calinski_harabasz_score, mutual_info_score, accuracy_score
```

```python
### Generate benchmark
P = np.array([[0.8, 0.2, 0.3],[0.2, 0.7, 0.4],[0.3, 0.4, 0.6]])
c = 3
n = 100
N = c*n
delta = 10
d = 2
dim = c*d
_,A,X = generate_benchmark(P,c,delta=delta,n=n,dim=d,sample_along_direction=True)
```

```python
### initialize variables
B = cp.Variable((c, c))
W = cp.Variable((N,c))
mu = cp.Variable((c,dim))
```

```python
net_div = cp.sum(cp.square(A-W@B@W.T),axis=1)
data_div = cp.sum(cp.multiply(W,cp.vstack([cp.hstack( [cp.norm(X[i,:]-mu[j,:]) for j in range(c)] ) for i in range(N)]) ),axis=1)
K = cp.vstack((net_div,data_div)).T
```

```python
cp.norm(K,1,axis=1).shape
```

```python
obj = cp.Minimize(cp.sum(cp.norm(K,1,axis=1)))
p = cp.Problem(obj)
p.solve()
```
